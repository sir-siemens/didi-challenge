```python
import pandas as pd
import numpy as np
import sys

# return a panda dataframe 
def ReadLabelFromOneFile(file_dir, index):
    # read index can ranges from 0 to 7480
    file_name = "/%06d.txt"%(index)
    print ( "Processing %s"%(file_name),  end='\r')
    absolute_file_name = file_dir + file_name
    d = np.genfromtxt(absolute_file_name, delimiter=' ', dtype=None)
    if len(d.shape) == 0:
        d.reshape(1)
        d = np.array([d])       
    df = pd.DataFrame(data = d)
    df.columns = ['class_type','truncation','occlusion','observe_angle','bb_x1','bb_y1','bb_x2','bb_y2','h','w','l','3Dx','3Dy','3Dz','yaw']
    # add image_index column
    sLength = len(df['class_type'])
    df['image_index'] = pd.Series( np.ones(sLength, dtype=int)*index, index=df.index)
    return df

def ReadLabelsFromAllFiles(file_dir, image_index_min=0, image_index_max=7480):
    # return a data frame from all the labels
    frames = []
    for i in range(image_index_min, image_index_max):
        df = ReadLabelFromOneFile(file_dir, i)
        # concatenate df to
        frames.append(df)
        merge = pd.concat(frames ,  ignore_index=True)
        merge['class_type'] = merge['class_type'].astype('str') 
        frames = [merge]
    print ('DONE')
    return frames[0]

```

### Read Labels from the files 

```python
df = ReadLabelsFromAllFiles(file_dir)
```
